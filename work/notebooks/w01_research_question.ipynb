```python
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

declining = (df["trend_direction"] == "down").mean()
stale_visible = ((df["days_since_last_update"] >= 180) & (df["impressions_90d"] >= 500)).mean()
low_ctr = ((df["impressions_90d"] >= 500) & (df["avg_position"] <= 20) & (df["ctr"] < 0.05)).mean()

print(f"Declining pages (trend_direction == down): {declining:.1%}")
print(f"Stale and visible pages (not updated in 180+ days, 500+ impressions): {stale_visible:.1%}")
print(f"Low CTR candidates (500+ impressions, top-20 position, CTR below 5%): {low_ctr:.1%}")
```

## 4. Careful words: what I can and can't claim

What this work can claim: we observed that certain signals such as staleness, impressions, position, and CTR are associated with pages flagged as declining. The ranked queue is directional and decision-support only. It helps a reviewer prioritize, it does not guarantee that refreshing a page will recover its traffic.
What this work cannot claim: that a refresh caused a recovery, that these signals prove a Google algorithm factor, or that any page is guaranteed to improve if acted on. All results are observed and measured, not causal.
