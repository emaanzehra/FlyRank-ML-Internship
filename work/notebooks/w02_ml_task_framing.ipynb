## 5. Why ML beats a fixed rule here

A fixed rule cannot combine multiple signals in a way that adapts to different page types. For example, a stale page with low impressions behaves differently from a stale page with high impressions and good position. A hand-written rule treats these the same. A learned model can find the interaction between staleness, impressions, position, CTR, and content age together, and weight them differently depending on the combination. The starter data already showed this: the random forest scored 3.1x better than the hand-written baseline at Precision@50 by learning these combinations from data rather than having them written in by hand.
